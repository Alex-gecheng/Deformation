```python
import trimesh
import pandas as pd
from scipy.interpolate import RBFInterpolator
```

```python

mesh = trimesh.load("plane.obj", force='mesh')
verts = mesh.vertices       
faces = mesh.faces           
print(mesh)
```

```python
df = pd.read_csv("point.csv")
print(df.head())

# 约束点
src_points = df[['x', 'y', 'z']].values
tar_points = df[['x1', 'y1', 'z1']].values
# print("Source Points:\n", src_points)
```

```python

rbf = RBFInterpolator(src_points, tar_points, kernel='cubic', neighbors=None,degree=0,)
```

```python
new_vertices = verts + rbf(verts)
```

```python
new_mesh = trimesh.Trimesh(vertices=new_vertices, faces=mesh.faces)
new_mesh.export("deformed.obj")

print("完成，已保存为 deformed.obj")
```
